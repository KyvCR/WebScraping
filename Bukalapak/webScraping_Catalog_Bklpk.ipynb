```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
```

```python
#get URL
url = ('https://www.bukalapak.com/c')
response = requests.get(url)
#print(response)

#check content
soup = bs(response.content, 'html.parser')
```

```python
catagories = soup.find('div',{'id' : 'categories'})

#get catagory product
header = []
for h3 in catagories.find_all('h3'):
  header.append(h3.text.strip())
header
```

```python
#get sub_catagory from catagory
catagory =[]
for all in catagories.find_all('div',{'class':'bl-flex-item masonry-item ut-category-item'}):
  x = []
  for h3 in all.find('h3'):
    x.append(h3.text.strip())
    list = []
    for a in all.find_all('a',{'class' : 'category-list ut-category-subcategory'}):
      list.append(a.text.strip())
    x.append(list)
  catagory.append(x)
catagory
```

```python
#inster into dataframe
df = pd.DataFrame(catagory, columns = ['Catagory', 'List'])
df
```

```python
df.to_csv('Katagori_Product_Bukalapak.csv', index=False)
```
