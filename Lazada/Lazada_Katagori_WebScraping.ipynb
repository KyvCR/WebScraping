```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests as rs
```

```python
#get url

url = ('https://www.lazada.co.id/')
response = rs.get(url)
response

soup = bs(response.content, "html.parser")
soup
```

```python
ctgp =  soup.find('div',{'class' : 'lzd-site-nav-menu-dropdown'})
ctgp
```

```python
#get main catalog
main_catalog = []
for main in ctgp.find_all('li',{'class' : 'lzd-site-menu-root-item'}):
  main_catalog.append(main.text.strip())
main_catalog
```

```python
#get sub main catalog
x = 0
sub_catalog = []
clas ='lzd-site-menu-sub Level_1_Category_No' #class name from sub catalog

#class name start from No 1 - No 12
while x < 12:
  x += 1
  for sub in ctgp.find_all('ul',{'class': f'{clas}{x}'}):
    z = [] 
    for span in sub.find_all('span'):
      z.append(span.text.strip())

  sub_catalog.append(z)
sub_catalog
```

```python
#get sub from sub catalog
sub_frm_sub = []
xx = 0
clas ='lzd-site-menu-sub Level_1_Category_No' #class name from sub catalog
while xx < 12:
  xx += 1
  for sub in soup.find_all('ul',{'class': f'{clas}{xx}'}): #looping  ul, dengan klas yang berbeda 
    sx = []
    for li in sub.find_all('li', {'class' : 'lzd-site-menu-sub-item'}):
      lix = []
      for sub_sub in li.find_all('li', {'class' : 'lzd-site-menu-grand-item'}):
        subx = []
        for a in sub_sub.find_all('a'):
            subx.append(a.text.strip())
        lix.append(subx)
      sx.append(lix)
  sub_frm_sub.append(sx)
sub_frm_sub
```

```python
print(len(main_catalog))
print(len(sub_catalog))
print(len(sub_frm_sub))
```

```python
df = pd.DataFrame({
    'main' : main_catalog,
    'SubList' : sub_catalog,
    'Sub catalog' : sub_frm_sub
})

df
```

```python
dict= []
lp = 0
for x in main_catalog:
  dict_catalog = {
      'CATALOG'   : x,
      'SUB CATALOG' : sub_catalog[lp],
      'Sub catalog' : sub_frm_sub[lp]
  }
  dict.append(dict_catalog)
  lp += 1
```
